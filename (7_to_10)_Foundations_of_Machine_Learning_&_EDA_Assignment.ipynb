## Analyze prices across airlines for a specific route

### Subtask:
Filter the data for a specific route (e.g., Delhi to Mumbai). Compare the average prices across different airlines for this route using descriptive statistics and visualizations (e.g., bar plot, box plot). Identify consistently cheaper or premium airlines.


**Reasoning**:
Filter the data for Delhi to Mumbai route, calculate average prices by airline, display the results, and create visualizations (bar plot and box plot) to compare prices across airlines for this route.



```python
# 1. Filter the data for Delhi to Mumbai route
df_delhi_mumbai = df_flight[(df_flight['source_city'] == 'Delhi') & (df_flight['destination_city'] == 'Mumbai')].copy()
print(f"Number of flights on the Delhi-Mumbai route: {len(df_delhi_mumbai)}")

# 2. Calculate the average price for each airline on this route
average_price_by_airline_delhi_mumbai = df_delhi_mumbai.groupby('airline')['price'].mean().sort_values()

# 3. Display the calculated average prices
print("\nAverage Price by Airline for Delhi-Mumbai Route:")
display(average_price_by_airline_delhi_mumbai)

# 4. Create a bar plot to visualize the average prices
plt.figure(figsize=(10, 6))
sns.barplot(x=average_price_by_airline_delhi_mumbai.index, y=average_price_by_airline_delhi_mumbai.values)
plt.title("Average Flight Price by Airline (Delhi-Mumbai Route)")
plt.xlabel("Airline")
plt.ylabel("Average Price")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 5. Create a box plot to visualize the distribution of prices
plt.figure(figsize=(10, 6))
sns.boxplot(x='airline', y='price', data=df_delhi_mumbai, order=average_price_by_airline_delhi_mumbai.index)
plt.title("Flight Price Distribution by Airline (Delhi-Mumbai Route)")
plt.xlabel("Airline")
plt.ylabel("Price")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 6. Identify consistently cheaper/premium airlines
print("\nAnalysis of Airline Prices on Delhi-Mumbai Route:")
print("Based on the average prices and price distribution:")
cheaper_airlines = average_price_by_airline_delhi_mumbai.head(2).index.tolist()
premium_airlines = average_price_by_airline_delhi_mumbai.tail(2).index.tolist()

print(f"- Consistently cheaper airlines: {cheaper_airlines}")
print(f"- Consistently premium airlines: {premium_airlines}")

print("\nPotential Reasons for Price Differences:")
print("- **Class of Service:** Some airlines primarily offer Economy class, while others offer Business or even First class, significantly impacting the average price.")
print("- **Service Quality and Amenities:** Premium airlines often offer better in-flight services, seat comfort, baggage allowances, and airport lounges.")
print("- **Route Demand and Competition:** Price can vary based on the demand for the specific route and the level of competition among airlines.")
print("- **Fleet and Aircraft Type:** Newer or more modern aircraft might have different operating costs and passenger capacities, influencing pricing.")
print("- **Branding and Market Positioning:** Airlines position themselves differently in the market, targeting different customer segments with corresponding pricing strategies.")
```
