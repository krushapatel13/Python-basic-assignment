2. How does broadcasting work in NumPy?

=> Broadcasting is a mechanism in NumPy that allows arithmetic operations to be performed on arrays of different shapes.
 When performing an operation between two arrays, NumPy compares their shapes element-wise starting from the trailing dimensions.
> Two dimensions are compatible when:
 - They are equal.
 - One of them is 1.

 If these conditions are not met, a ValueError is raised, indicating that the arrays are not compatible for broadcasting.

If the dimensions are compatible, the smaller array is "broadcast" across the larger array so that they have compatible shapes for the operation. This broadcasting does not actually create copies of the smaller array; it's a conceptual expansion.

8. What does the term “vectorized operation” mean in NumPy?

=>  In NumPy, a "vectorized operation" refers to performing mathematical operations on entire arrays at once, rather than element by element using Python loops. These operations are implemented in C and are significantly faster and more efficient than their pure Python counterparts. Essentially, instead of writing a for loop to add two arrays element-wise, you can simply use the + operator directly on the NumPy arrays, and NumPy handles the element-wise addition internally in an optimized way. This eliminates the overhead of Python loops, leading to substantial performance gains, especially for large datasets.

17. Explain the difference between apply() and map() in Pandas?

=> In Pandas, both apply() and map() are used to apply a function to elements of a Series or DataFrame, but they have key differences in their application and flexibility:

>map():

- Primarily used for element-wise transformation on a Series.
- Takes a dictionary, Series, or function as input.
- When a dictionary or Series is used, it performs a lookup and replaces the values.
- When a function is used, it applies the function to each element of the Series.
- Is generally faster than apply() for element-wise operations on a Series, especially with a dictionary or Series lookup.

>apply():

- Can be used on both Series and DataFrames.
- When used on a Series, it applies a function to each element or to the entire Series.
- When used on a DataFrame, it can apply a function along an axis (rows or columns).
- Is more flexible than map() as it can operate on entire rows or columns, not just element-wise.
- Can return a Series, a DataFrame, or a scalar value.
- Can be slower than map() for element-wise operations due to its greater flexibility.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

=> NumPy's array slicing is significantly faster than Python's list slicing due to several underlying differences in how they are implemented:

- Contiguous Memory Allocation: NumPy arrays are stored in a contiguous block of memory. This allows NumPy to quickly access and slice data using efficient pointer arithmetic without needing to access elements individually like Python lists. Python lists, on the other hand, store references to objects that can be scattered throughout memory, making slicing involve more overhead to follow these pointers.
- Homogeneous Data Type: NumPy arrays require all elements to be of the same data type. This allows NumPy to work with fixed-size memory blocks and leverage highly optimized, type-specific C code for operations like slicing. Python lists can hold elements of different data types, which requires more flexibility but comes with performance costs due to type checking and dynamic sizing.
- No Overhead for Individual Objects: When slicing a Python list, a new list is created, and each element is copied. For large lists, this copying process can be slow. NumPy slicing, in contrast, often creates a "view" into the original array's memory. This view is a new array object that refers to the same data, avoiding the overhead of copying elements. This makes NumPy slicing very fast, especially for large arrays.
Optimized C Implementation: NumPy's core operations, including slicing, are implemented in C. This low-level implementation allows for much faster execution compared to Python's interpreted code for list slicing.

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Create a sample DataFrame
np.random.seed(0)
data = {'col1': np.random.rand(100),
        'col2': np.random.rand(100) * 2,
        'col3': np.random.rand(100) * 0.5 + 5}
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
