### Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

-> Anomaly Detection, also known as outlier detection, is the process of identifying data points, events, or observations that deviate significantly from the norm. These 'anomalies' or 'outliers' can often indicate critical incidents like fraud, system malfunctions, or rare events.

Here's a breakdown of its types with examples:

Types of Anomalies:

1. Point Anomalies: These are individual data points that are abnormal relative to the rest of the data. They are single instances that stand out.

* Example:
 A single transaction of 10,000 from a credit card that usually has transactions under $$10,000 from a credit card that usually has transactions under $100

2. Contextual Anomalies: A data instance is considered anomalous in a specific context, but not otherwise. The context could be temporal, spatial, or behavioral.
* Example:
A temperature reading of 30°C in the desert in July might be normal, but the same temperature reading in Antarctica in January would be a contextual anomaly. Another example could be a sudden high volume of website traffic at 3 AM for a business that typically operates during the day; while high traffic isn't an anomaly on its own, its occurrence at 3 AM makes it contextual.
3. Collective Anomalies: A collection of related data instances are anomalous with respect to the entire dataset, even if individual data points within the collection are not anomalous.
* Example: A sudden drop in website traffic for a few consecutive hours might not be individually anomalous if each hour's traffic is within a 'normal' range, but collectively, the sustained low traffic could indicate a server issue or a denial-of-service attack.

### Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.


-> Isolation Forest

- Approach: Isolation Forest works by isolating anomalies rather than profiling normal data points. It does this by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of that selected feature. Anomalies are data points that require fewer splits to be isolated compared to normal data points. It builds an ensemble of isolation trees (iTrees).
- Suitable Use Cases: Highly effective for high-dimensional datasets and when dealing with a large number of normal instances and a small number of anomalies. It's particularly good for large datasets because of its linear time complexity. Ideal for fraud detection, network intrusion detection, and other scenarios where anomalies are rare and distinct.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

- Approach: DBSCAN is a density-based clustering algorithm that groups together closely packed data points, marking as outliers those points that lie alone in low-density regions. It requires two parameters: eps (epsilon), which defines the maximum distance between two samples for one to be considered as in the neighborhood of the other, and min_samples, the number of samples (or total weight) in a neighborhood for a point to be considered as a core point.
- Suitable Use Cases: Excellent for datasets with clusters of varying shapes and sizes, and for detecting outliers that do not belong to any cluster. It does not require specifying the number of clusters beforehand. Suitable for geographical data analysis, image processing, and identifying anomalies in spatial data where density plays a crucial role.

Local Outlier Factor (LOF)

- Approach: LOF measures the local deviation of density of a given data point with respect to its neighbors. It considers as outliers those samples that have a substantially lower density than their neighbors. The 'local' aspect means that it only compares the density of a point to that of its neighbors, making it effective for datasets where the anomaly score can vary depending on the region of the data space.
- Suitable Use Cases: Very effective for detecting local outliers in datasets where different regions may have different densities. It works well when anomalies might be normal in a different part of the dataset, but anomalous in their specific neighborhood. Useful in manufacturing defect detection, medical diagnosis, and other applications where the definition of an outlier is relative to the immediate surroundings.

### Question 3: What are the key components of a Time Series? Explain each with one example

-> A Time Series is a sequence of data points indexed in time order. Most commonly, a time series is a sequence taken at successive equally spaced points in time. The key components that describe the characteristics of a time series are Trend, Seasonality, Cyclical, and Irregular/Random Component.

Here's an explanation of each with an example:

1. Trend: This refers to the long-term increase or decrease in the data over time. It represents the underlying direction of the data.

Example: The increasing average global temperature over the past century indicates an upward trend in climate data.

2. Seasonality: This component refers to predictable and recurring patterns or cycles within a fixed period, such as a day, week, month, or year. These patterns are typically driven by calendar-related events.

Example: Retail sales often show a seasonal pattern, with peaks during holiday seasons (e.g., Christmas) and dips during off-peak months. Daily electricity consumption also shows seasonality, with higher usage during working hours and lower usage at night.

3. Cyclical: These are patterns that describe fluctuations around the trend, usually lasting for more than one year, and are often irregular in length and magnitude. Unlike seasonality, cyclical variations are not fixed and do not have a constant period.

Example: Economic recession and recovery cycles, which can last for several years. For instance, a period of economic growth followed by a downturn, then another period of growth, would represent a cyclical component.

4. Irregular/Random Component (Noise): This component represents the unpredictable and unsystematic fluctuations in the data that cannot be explained by trend, seasonality, or cyclical patterns. It's often referred to as 'noise'.

Example: A sudden, unexpected surge in stock prices due to an unforeseen geopolitical event or a one-time company announcement would be considered an irregular component. Similarly, a random power outage briefly affecting sales data would fall into this category.

### Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?


->
A time series is said to be stationary if its statistical properties—such as mean, variance, and autocorrelation—are all constant over time. This means that the series does not exhibit trends, seasonal patterns, or other systematic changes that would make its statistical characteristics dependent on the time at which the observations are made.

Why is stationarity important?

Many time series models (e.g., ARIMA) assume that the underlying data generation process is stationary. Non-stationary series can lead to spurious regressions and unreliable forecasts.

- How to Test for Stationarity:


There are two main approaches to testing for stationarity:

1. Visual Inspection (Plotting):

- Time Plot: Plot the time series data over time. A stationary series will tend to hover around a constant mean with constant variance. Look for trends (upward or downward slopes), seasonality (repeating patterns), or changes in variance (e.g., increasing fluctuations over time).


- Autocorrelation Function (ACF) Plot: For a stationary series, the ACF typically drops off quickly to zero. For a non-stationary series (especially one with a trend), the ACF often decays very slowly or remains high for many lags.
2. Statistical Hypothesis Tests (Unit Root Tests):

- Augmented Dickey-Fuller (ADF) Test: This is one of the most common statistical tests for stationarity. The null hypothesis (H0) is that the time series has a unit root (i.e., it is non-stationary). The alternative hypothesis (H1) is that the series is stationary. A small p-value (typically < 0.05) leads to the rejection of the null hypothesis, suggesting stationarity.
- Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test: The KPSS test is complementary to the ADF test. Its null hypothesis (H0) is that the time series is trend-stationary (or level-stationary). The alternative hypothesis (H1) is that the series is non-stationary. A small p-value (typically < 0.05) leads to the rejection of the null hypothesis, suggesting non-stationarity.

How to Transform a Non-Stationary Series into a Stationary One:

If a time series is found to be non-stationary, it can often be transformed into a stationary one using various techniques:

1. Differencing: This is the most common technique. It involves computing the difference between consecutive observations. First-order differencing (Y_t - Y_{t-1}) can often remove a linear trend. If there's a strong seasonal component, seasonal differencing (Y_t - Y_{t-L}), where L is the length of the season, can be applied.

Example: If your data shows an upward trend, df['diff_sales'] = df['sales'].diff().dropna() would create a new series that is often more stationary.

2. Log Transformation: If the variance of the series is not constant over time (i.e., it increases with the mean), a logarithmic transformation (log(Y_t)) can help stabilize the variance.

Example: df['log_sales'] = np.log(df['sales'])

3. Square Root or Cube Root Transformation: Similar to log transformation, these can also help stabilize variance or make the series more Gaussian-like.

Example: df['sqrt_sales'] = np.sqrt(df['sales'])

4. Decomposition: Time series can be decomposed into trend, seasonal, and residual components. Once decomposed, the trend and seasonal components can be removed, leaving a more stationary residual component.

Example: Using seasonal_decompose from statsmodels to extract and then analyze the residual component.

### Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.


-> 1. AR (AutoRegressive) Model


- Structure: An AR model of order p (denoted as AR(p)) predicts the future value of a variable based on a linear combination of its past p values. It assumes that the current value is a function of previous values plus a random error term. The 'auto' in autoregressive refers to the regression of the variable on itself (its past values).
- Equation: Y_t = c + φ₁Y_{t-1} + φ₂Y_{t-2} + ... + φₚY_{t-p} + ε_t

Y_t: Current value

c: Constant

φ₁...φₚ: AutoRegressive coefficients

ε_t: White noise error term

- Application: Best suited for time series data that exhibits a clear dependence on its past values, such as stock prices, temperature readings, or economic indicators that show momentum.
2. MA (Moving Average) Model
- Structure: An MA model of order q (denoted as MA(q)) predicts the future value of a variable based on a linear combination of past q error terms (residuals). It assumes that the current value is a function of current and past error terms, implying that shocks or random disturbances have a lasting effect on the series.
- Equation: Y_t = μ + ε_t + θ₁ε_{t-1} + θ₂ε_{t-2} + ... + θ_qε_{t-q}

Y_t: Current value

μ: Mean of the series

ε_t: Current white noise error term

θ₁...θ_q: Moving Average coefficients for past error terms

- Application: Effective for time series where fluctuations are primarily due to random shocks that dissipate over time, like short-term demand fluctuations or white noise processes.
3. ARIMA (AutoRegressive Integrated Moving Average) Model
- Structure: ARIMA(p, d, q) combines AR and MA components with an 'Integrated' (I) component. The 'I' refers to differencing the time series d times to make it stationary. Once differenced, the AR(p) and MA(q) components are applied to the stationary series.

p: Order of the AutoRegressive part (number of past observations)

d: Order of differencing (number of times the raw observations are differenced)

q: Order of the Moving Average part (number of past error terms)

- Application: Widely used for forecasting non-stationary time series data that can be made stationary by differencing. Common in economics, finance, and engineering for sales forecasting, stock market prediction, and process control.

4. SARIMA (Seasonal AutoRegressive Integrated Moving Average) Model
- Structure: SARIMA(p, d, q)(P, D, Q)m extends ARIMA by adding a seasonal component. It handles both non-seasonal and seasonal patterns. The seasonal part of the model is represented by (P, D, Q)m:

P: Seasonal AutoRegressive order

D: Seasonal differencing order

Q: Seasonal Moving Average order

m: Number of observations per season (e.g., 12 for monthly data, 4 for quarterly data)

- Application: Essential for time series with clear seasonal patterns in addition to non-seasonal trends and fluctuations. Examples include retail sales (monthly/quarterly cycles), utility consumption (daily/weekly/yearly cycles), and tourism data.

5. SARIMAX (Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors) Model
- Structure: SARIMAX(p, d, q)(P, D, Q)m with X extends SARIMA by allowing the inclusion of exogenous variables (also known as covariates). These are external variables that are not part of the time series itself but can influence its behavior.

It incorporates the SARIMA structure for the time series's inherent patterns.

X: Represents one or more external variables that can be included in the model's regression component.

- Application: Used when external factors are known to impact the time series and can improve forecasting accuracy. For example, predicting sales (Y_t) might use SARIMA components along with advertising spend or competitor prices (X) as exogenous variables. Other applications include energy demand forecasting (with temperature as an exogenous variable) or stock market prediction (with economic indicators).

### Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs

# 1. Generate a synthetic 2D dataset with some anomalies
# Generate 'normal' data points
X, _ = make_blobs(n_samples=300, centers=1, cluster_std=0.6, random_state=42)

# Add some 'anomalous' data points
anomalies = np.array([
    [4, 4], [3.5, -2], [-3, 3.5], [-2, -4], [0, 5],
    [5, 0], [-5, 0], [0, -5], [6, 6], [-6, -6]
])

X_combined = np.vstack([X, anomalies])

# Create a DataFrame for better handling
df = pd.DataFrame(X_combined, columns=['Feature1', 'Feature2'])

print(f"Generated dataset shape: {df.shape}")

# 2. Apply Isolation Forest
# You can adjust parameters like 'contamination' based on your domain knowledge
# contamination: The proportion of outliers in the dataset. Default is 'auto'.
# Setting a specific value helps the model to expect a certain percentage of outliers.
# For this synthetic dataset, we know the approximate contamination.
model = IsolationForest(n_estimators=100, contamination=0.05, random_state=42, verbose=0, n_jobs=-1)
model.fit(df[['Feature1', 'Feature2']])

# Predict the anomaly scores (lower score indicates higher anomaly likelihood)
df['anomaly_score'] = model.decision_function(df[['Feature1', 'Feature2']])

# Predict the anomaly labels (-1 for anomalies, 1 for normal)
df['anomaly_label'] = model.predict(df[['Feature1', 'Feature2']])

# Separate normal and anomalous points for plotting
anomalous_points = df[df['anomaly_label'] == -1]
normal_points = df[df['anomaly_label'] == 1]

# 3. Visualize the anomalies on a 2D scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x=normal_points['Feature1'], y=normal_points['Feature2'], color='blue', label='Normal Points', s=50, alpha=0.7)
sns.scatterplot(x=anomalous_points['Feature1'], y=anomalous_points['Feature2'], color='red', label='Anomalies', s=100, marker='X', edgecolor='black')

plt.title('Anomaly Detection using Isolation Forest (Synthetic Data)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

print("\nDetected Anomalies:")
print(anomalous_points[['Feature1', 'Feature2', 'anomaly_score']])
```

### Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
(Include your Python code and output in the code box below.)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs

# 1. Generate a synthetic 2D dataset with local anomalies

# Generate some 'normal' data clusters
X1, _ = make_blobs(n_samples=150, centers=[[2, 2]], cluster_std=0.5, random_state=42)
X2, _ = make_blobs(n_samples=100, centers=[[-2, -2]], cluster_std=0.8, random_state=42)
X3, _ = make_blobs(n_samples=80, centers=[[0, 3]], cluster_std=0.3, random_state=42)

# Add some 'anomalous' data points (local outliers and isolated points)
anomalies = np.array([
    [6, 6], [5.5, 0], [-5, -5], [0, -5], [3, -3],
    [0, 0], [4, -1], [-1, 5]
])

X_combined = np.vstack([X1, X2, X3, anomalies])

# Create a DataFrame for better handling
df_lof = pd.DataFrame(X_combined, columns=['Feature1', 'Feature2'])

print(f"Generated dataset shape: {df_lof.shape}")

# 2. Apply Local Outlier Factor (LOF)
# n_neighbors: Number of neighbors to consider for the local density. Typical values are 20 to 100.
# contamination: The proportion of outliers in the dataset. Default is 'auto'.
# A negative LOF score typically indicates an inlier, and positive indicates an outlier.
# Values closer to 1 are normal, values significantly higher than 1 are anomalies.
model_lof = LocalOutlierFactor(n_neighbors=20, contamination='auto', novelty=False) # novelty=False for unsupervised learning

# Fit the model and predict outlier scores
# The fit_predict method returns -1 for outliers and 1 for inliers
df_lof['anomaly_label'] = model_lof.fit_predict(df_lof[['Feature1', 'Feature2']])

# Get the negative outlier factor. Higher absolute value means more anomalous.
# LOF returns 'negative_outlier_factor_'. So, -LOF_score gives a positive value for outlierness.
# Lower LOF scores indicate higher anomaly likelihood (higher negative_outlier_factor_ value).
# For visualization, we often want a score where higher = more anomalous. So we invert it.
raw_scores = model_lof.negative_outlier_factor_
df_lof['lof_score'] = -raw_scores # Invert to get a more intuitive 'outlierness' score


# Separate normal and anomalous points for plotting based on the predicted label
anomalous_points_lof = df_lof[df_lof['anomaly_label'] == -1]
normal_points_lof = df_lof[df_lof['anomaly_label'] == 1]

# 3. Visualize the anomalies on a 2D scatter plot
plt.figure(figsize=(12, 9))

# Plot normal points
sns.scatterplot(
    x=normal_points_lof['Feature1'],
    y=normal_points_lof['Feature2'],
    color='blue',
    label='Normal Points',
    s=50, # size of points
    alpha=0.7
)

# Plot anomalous points
sns.scatterplot(
    x=anomalous_points_lof['Feature1'],
    y=anomalous_points_lof['Feature2'],
    hue=anomalous_points_lof['lof_score'], # Color by LOF score
    size=anomalous_points_lof['lof_score'], # Size by LOF score
    sizes=(100, 400), # Range of sizes for anomalies
    palette='viridis_r', # Reverse viridis to make higher scores darker/more distinct
    marker='X',
    edgecolor='black',
    # Removed conflicting 'label' argument
    legend='brief'
)

plt.title('Anomaly Detection using Local Outlier Factor (LOF)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()

print("\nDetected Anomalies (Top 10 by LOF score):")
print(df_lof.sort_values(by='lof_score', ascending=False).head(10)[['Feature1', 'Feature2', 'lof_score', 'anomaly_label']])
```

#### 1. Suitability of Isolation Forest and Local Outlier Factor (LOF) for Real-time Streaming Data

For real-time, streaming 15-minute energy consumption data, both Isolation Forest and Local Outlier Factor (LOF) are strong candidates for anomaly detection due to their inherent characteristics:

**Isolation Forest:**
*   **Efficiency for High-Dimensional Data:** While energy usage might initially seem low-dimensional, when incorporating features like region, weather conditions (temperature, humidity, cloud cover), time-of-day, day-of-week, and historical usage patterns, the data becomes multi-dimensional. Isolation Forest handles high-dimensional data efficiently, making it suitable for this rich feature set.
*   **Scalability:** It has a linear time complexity, which is crucial for large, continuous streams of data. It doesn't need to compute distances between all data points, making it faster than many other methods.
*   **Doesn't Profile Normal Data:** Instead of building a model of normal behavior, it directly isolates anomalies. This is advantageous because defining 'normal' in a dynamic energy grid can be complex and ever-changing. Anomalies are points that are easier to separate from the rest of the data.

## Explain Short-term Forecasting Model

### Justification for SARIMAX Model Selection in Short-term Energy Demand Forecasting

For short-term energy demand forecasting with 15-minute consumption data, considering seasonality and exogenous variables like weather conditions, the **SARIMAX (Seasonal AutoRegressive Integrated Moving Average with eXogenous regressors) model** is an exceptionally suitable choice. Here's a detailed justification:

1.  **Why SARIMAX is Chosen: Handling Seasonal Patterns and External Influences**
    *   **Comprehensive Seasonality Capture:** Energy demand exhibits strong seasonal patterns. At a 15-minute frequency, we expect not only daily and weekly seasonality but also potentially yearly seasonality. SARIMAX explicitly incorporates both non-seasonal (ARIMA) and seasonal components (Seasonal ARIMA) to capture these recurring patterns. This is crucial as energy demand fluctuates predictably based on time of day (e.g., peak demand during working hours), day of week (e.g., lower demand on weekends), and time of year (e.g., higher demand in summer for cooling or winter for heating).
    *   **Exogenous Variable Integration:** A key advantage of SARIMAX over simpler models like ARIMA or SARIMA is its ability to include **eXogenous regressors (X)**. Weather conditions (temperature, humidity, cloud cover) are primary drivers of energy demand. By integrating these external variables into the model, SARIMAX can account for their direct impact on consumption, leading to significantly improved forecasting accuracy. For example, a sudden heatwave will increase cooling demand, and SARIMAX can model this relationship.
    *   **Handling Non-Stationarity:** Energy demand data often exhibits trends (e.g., increasing overall consumption due to population growth) and changing variance, making it non-stationary. The 'I' (Integrated) components in both the non-seasonal and seasonal parts of SARIMAX allow for differencing the data to achieve stationarity, a prerequisite for many time series models, without losing information about trends or seasonality.

2.  **Impact of 15-Minute Data Frequency on Modeling Approach**
    *   **High Granularity, Multiple Seasonalities:** The 15-minute frequency provides a very granular view of energy consumption. This high resolution means we need to meticulously identify and model multiple seasonal components:
        *   **Daily Seasonality:** 96 observations per day (24 hours * 4 15-minute intervals). This pattern reflects human activity cycles within a single day.
        *   **Weekly Seasonality:** 672 observations per week (7 days * 96 observations/day). This pattern captures differences in demand between weekdays and weekends.
        *   **Potentially Yearly Seasonality:** While 15-minute data might make yearly seasonality (m = 35040 observations per year) computationally intensive to model directly with a single SARIMAX 'm' parameter, it can be captured through the trend component and interaction with exogenous weather variables. For very long seasonal periods, alternative approaches like Fourier series might be considered alongside SARIMAX, or the 'm' parameter can be set to represent a daily or weekly cycle, with longer-term seasonality implicitly handled by the exogenous variables and the non-seasonal components.
    *   **Increased Data Volume:** High-frequency data generates a large volume of observations, which can be computationally demanding. However, modern time series libraries are optimized for this, and the detailed patterns captured at this frequency are essential for precise short-term forecasts.

3.  **Role of Exogenous Variables (Weather Conditions)**
    *   **Direct Causal Relationship:** Weather variables like ambient temperature, humidity, and possibly solar radiation (for solar generation integration) have a direct and strong causal relationship with energy demand. High temperatures increase cooling load, while low temperatures increase heating load. Humidity can exacerbate the perception of temperature, and cloud cover affects solar power generation and hence net demand.
    *   **Improving Accuracy:** By including these as exogenous regressors, SARIMAX can dynamically adjust forecasts based on predicted weather. This is critical for short-term predictions (e.g., next few hours to next day) where weather changes can be significant and rapid. Without them, the model would only learn the average seasonal pattern, leading to large forecast errors during unusual weather events.
    *   **Feature Engineering:** It might be beneficial to engineer additional weather-related features, such as heating degree days (HDD), cooling degree days (CDD), or lagged weather variables, to capture more nuanced relationships.

4.  **Specific Components of SARIMAX and Energy Demand Patterns**
    *   **AR(p) - AutoRegressive:** Captures the dependency of current demand on past demand values. For instance, high demand in the previous 15-minute interval might indicate high demand in the current interval due to inertia in system state or consumer behavior.
    *   **I(d) - Integrated (Differencing):** Addresses non-stationarity in the mean. First-order differencing (d=1) can remove linear trends. If a 15-minute series shows a general upward or downward trend over several days/weeks, differencing helps stabilize the mean.
    *   **MA(q) - Moving Average:** Accounts for the dependency of current demand on past forecast errors (shocks). This helps capture short-term random fluctuations that are not explained by past demand values.
    *   **Seasonal AR(P) - Seasonal AutoRegressive:** Captures the dependency of current demand on demand from *previous seasons*. For 15-minute data, this would mean comparing current demand to demand at the same time yesterday (P=1, m=96) or the same time last week (P=1, m=672). This is crucial for daily and weekly patterns.
    *   **Seasonal I(D) - Seasonal Integrated (Seasonal Differencing):** Addresses non-stationarity in seasonal patterns. If the seasonal pattern itself changes over time (e.g., the daily peak demand increases each year), seasonal differencing helps stabilize this seasonal component.
    *   **Seasonal MA(Q) - Seasonal Moving Average:** Captures dependency on past seasonal forecast errors. This can account for recurring but unmodeled shocks that happen at specific seasonal intervals.
    *   **X - Exogenous Regressors:** As discussed, this includes variables like temperature, humidity, day of week, hour of day, and holiday indicators. These variables directly influence energy consumption and their inclusion significantly enhances the model's predictive power by explaining variations not captured by the time-based components alone.

## Explain Validation and Monitoring

### Subtask:

#### 1. Validation Approach for Anomaly Detection Model

Validating an anomaly detection model, especially in a real-time streaming environment, can be challenging due to the inherent rarity of anomalies and often the lack of ground truth labels. The approach typically involves a combination of supervised and unsupervised techniques.

*   **Metrics (if labeled data is available for a training phase):**
    *   **Precision:** Measures the proportion of correctly identified anomalies out of all instances flagged as anomalies (True Positives / (True Positives + False Positives)). High precision means fewer false alarms, which is crucial in operational settings to avoid alert fatigue.
    *   **Recall:** Measures the proportion of actual anomalies that were correctly identified (True Positives / (True Positives + False Negatives)). High recall means fewer missed anomalies, important for critical events like system malfunctions.
    *   **F1-score:** The harmonic mean of precision and recall (2 * (Precision * Recall) / (Precision + Recall)). It provides a single metric that balances both.
    *   **False Positive Rate (FPR):** The proportion of normal instances incorrectly classified as anomalies. Minimizing FPR is often a key objective.
    *   **False Negative Rate (FNR):** The proportion of actual anomalies incorrectly classified as normal. Minimizing FNR is critical for high-risk anomalies.
    *   **Receiver Operating Characteristic (ROC) Curve and Area Under the Curve (AUC):** Useful for evaluating the model's ability to distinguish between normal and anomalous classes across various threshold settings. A higher AUC indicates better discriminatory power.

*   **Handling the lack of ground truth in real-time streaming data:**
    *   **Expert Feedback Loop:** Implement a system where flagged anomalies are reviewed by domain experts (e.g., power grid operators). Their feedback (confirming or denying an anomaly) can serve as ground truth for continuous learning and model refinement.
    *   **Sampling and Manual Labeling:** Periodically sample the data and have a subset manually labeled by experts. This can help in generating limited ground truth for evaluating model performance over time.
    *   **Synthetic Anomalies:** Inject known synthetic anomalies into the data stream to test the model's detection capabilities. This helps benchmark performance against specific types of anomalies.
    *   **Unsupervised Metrics:** In the absence of labels, evaluate the consistency of anomaly scores. For instance, do known 'normal' periods consistently yield low anomaly scores? Are sudden shifts in score distribution detected?
    *   **Clustering of Anomalies:** Analyze the characteristics of detected anomalies. Do they group into meaningful patterns that suggest real-world events? This can provide qualitative validation.

#### 2. Validation Approach for Short-Term Forecasting Model

Validating forecasting models involves assessing how well the model predicts future values that it has not seen before. This is typically done using out-of-sample data.

*   **Metrics:**
    *   **Mean Absolute Error (MAE):** The average of the absolute differences between forecasted values and actual values. It's easy to interpret as the average magnitude of errors.
    *   **Root Mean Squared Error (RMSE):** The square root of the average of the squared differences. It gives higher weight to larger errors, making it sensitive to outliers.
    *   **Mean Absolute Percentage Error (MAPE):** Expresses the error as a percentage of the actual value. Useful for comparing forecast accuracy across different scales of data, but can be problematic if actual values are zero or very close to zero.
    *   **Symmetric Mean Absolute Percentage Error (sMAPE):** Addresses some limitations of MAPE by using the average of the absolute actual and forecasted values in the denominator.
    *   **R-squared (Coefficient of Determination):** Measures the proportion of the variance in the dependent variable that is predictable from the independent variables. While less common for time series forecasting errors than for regression, it can provide context on model fit.

*   **Out-of-sample Forecasting and Backtesting:**
    *   **Hold-out Validation:** Split the historical data into a training set and a validation set (e.g., last 12 months for monthly data). Train the model on the training set and evaluate its performance on the validation set.
    *   **Rolling Forecast Origin (Walk-Forward Validation):** This simulates a real-time forecasting scenario. The model is trained on data up to a certain point, forecasts the next period, then the training window is advanced to include the actual value of that period, and the process repeats. This provides a robust estimate of performance over time and helps assess model stability.
    *   **Backtesting:** Apply the model to past data where actual outcomes are known, and compare predictions against these outcomes. This helps identify periods where the model performed well or poorly and understand the conditions under which it operates effectively.
    *   **Cross-Validation (Time Series Specific):** Techniques like `TimeSeriesSplit` in scikit-learn allow for splitting time series data into multiple train-test pairs while preserving the temporal order, preventing data leakage from the future into the past.

#### 3. Continuous Monitoring of Both Models

Real-time monitoring is crucial to ensure that models maintain their performance as data patterns evolve and new anomalies emerge.

*   **Monitoring Dashboards:** Develop interactive dashboards (e.g., using Grafana, Kibana, custom web apps) that display key performance indicators (KPIs) for both models.
    *   **For Anomaly Detection:**
        *   Number of anomalies detected per hour/day.

        *   Trend of false positives/negatives (if ground truth feedback is incorporated).
        *   Alerts for sudden changes in detection rate.
    *   **For Forecasting:**
        *   Real-time comparison of forecasts vs. actuals.
        *   Rolling MAE/RMSE/MAPE metrics over recent periods.
        *   Visualization of forecast errors and their distribution.
        *   Alerts for significant deviations (e.g., actual value outside of forecast confidence intervals for multiple consecutive periods).

*   **Anomaly Detection on Model Residuals (Forecast Errors):**
    *   Apply anomaly detection techniques (like Isolation Forest or LOF) directly to the forecast errors. If forecast errors themselves start exhibiting anomalous behavior (e.g., consistently large errors, errors showing a pattern), it indicates that the forecasting model might be degrading or that new underlying patterns are emerging.
    *   This provides an early warning system for forecast model degradation.

*   **Tracking Changes in Model Performance Metrics Over Time:**
    *   Store historical performance metrics (MAE, RMSE, Precision, Recall) in a time series database.
    *   Monitor these metrics for trends or sudden drops/spikes. For example, a sustained increase in MAE or a drop in precision indicates model drift or concept drift in the underlying data.
    *   Set up statistical process control (SPC) charts or control limits on these metrics to automatically flag when performance falls outside acceptable bounds.

#### 4. Real-time Adaptation and Retraining Strategies

Models, especially in dynamic environments like power grids, need to adapt to evolving patterns and conditions.

*   **Scheduled Retraining:**
    *   **Frequency:** Retrain models periodically (e.g., daily, weekly, monthly) using the latest available data. The optimal frequency depends on the rate of concept drift and computational resources.
    *   **Windowing:** Use a moving window of recent data for retraining to capture the most current patterns while discarding very old, potentially irrelevant data.

*   **Event-Driven Retraining (due to concept drift or performance degradation):**
    *   **Trigger Conditions:** Retraining is triggered when monitoring dashboards or anomaly detection on residuals indicate a significant degradation in model performance (e.g., MAE exceeds a threshold, high FPR/FNR persists, model's residual anomaly detector flags errors).
    *   **Performance Monitoring:** Continuous monitoring of metrics as described above directly informs when retraining is necessary.
    *   **Data Drift Detection:** Monitor statistical properties of incoming data (e.g., mean, variance, distribution of features) for significant shifts compared to the training data. Data drift often precedes concept drift and can be an early trigger for retraining.

*   **Deployment of Model Updates without Disrupting Live System:**
    *   **Shadow Deployment (A/B Testing):** Deploy the new model alongside the old one. Route a small percentage of real-time data to the new model (in shadow mode, meaning its predictions are not used for live decisions but are logged). Compare its performance against the old model. Once confidence is high, gradually increase traffic to the new model.
    *   **Canary Deployments:** Gradually roll out the new model to a small subset of the production environment or specific regions. Monitor its performance closely, and if successful, expand the rollout. If issues arise, quickly revert to the old model.
    *   **Blue-Green Deployment:** Maintain two identical production environments. One (blue) serves live traffic, while the other (green) holds the new model. Once the new model is validated in the green environment, traffic is switched over instantaneously. This allows for quick rollback if necessary.
    *   **Model Versioning:** Implement robust model versioning and management to track which model version is active, its training data, and its performance history. This ensures traceability and facilitates rollbacks.

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings

# Suppress warnings that might arise from model convergence issues or statistical tests
warnings.filterwarnings("ignore")

# --- 1. Conceptual function to simulate real-time 15-minute data ingestion ---

def simulate_streaming_data(start_time, current_time, region_id):
    """
    Simulates a single 15-minute data point for energy usage, temperature, and humidity.
    In a real-world scenario, this data would come from sensors, APIs, or a message queue.
    """
    # Simulate energy usage with a base value, some trend, and random noise
    base_usage = 100 + (current_time - start_time).days * 0.1 # Simple upward trend
    seasonal_usage = 20 * np.sin(current_time.hour * np.pi / 12) # Daily seasonality
    noise = np.random.normal(0, 5)
    energy_usage = max(0, base_usage + seasonal_usage + noise)

    # Simulate weather conditions (temperature, humidity)
    # These could also have daily/seasonal patterns and be influenced by actual weather data
    temperature = 20 + 5 * np.sin(current_time.timetuple().tm_yday * 2 * np.pi / 365) + np.random.normal(0, 1)
    humidity = 60 + 10 * np.sin(current_time.timetuple().tm_yday * 2 * np.pi / 365) + np.random.normal(0, 2)

    data_point = {
        'timestamp': current_time,
        'region': f'Region_{region_id}',
        'energy_usage': energy_usage,
        'temperature': temperature,
        'humidity': humidity
    }
    # In a real system, this would be pushed to a message queue (Kafka, RabbitMQ) or a real-time database
    return data_point


# --- 2. Conceptual AnomalyDetector class ---

class AnomalyDetector:
    def __init__(self, model_type='IsolationForest', contamination='auto'):
        """
        Initializes the Anomaly Detector with a specified model.
        In a real system, the model parameters might be tuned dynamically or loaded from a config.
        """
        if model_type == 'IsolationForest':
            self.model = IsolationForest(contamination=contamination, random_state=42)
        # elif model_type == 'LOF':
        #     from sklearn.neighbors import LocalOutlierFactor
        #     self.model = LocalOutlierFactor(contamination=contamination, novelty=True) # novelty=True for prediction on new data
        # elif model_type == 'DBSCAN':
        #     # DBSCAN is primarily a clustering algorithm; anomaly detection is a byproduct
        #     # It would require fitting on existing data and then checking if new points are noise
        #     # This is more complex for direct 'predict' on single new points compared to IsolationForest/LOF
        #     raise NotImplementedError("DBSCAN is not straightforward for real-time single-point anomaly prediction.")
        else:
            raise ValueError("Unsupported anomaly detection model type.")

        self.trained = False
        self.features_for_detection = ['energy_usage', 'temperature', 'humidity'] # Define relevant features

    def train(self, historical_data: pd.DataFrame):
        """
        Trains the anomaly detection model on historical data.
        In a real system, this would happen periodically or upon significant data distribution shifts.
        """
        if not historical_data.empty:
            X_train = historical_data[self.features_for_detection]
            self.model.fit(X_train)
            self.trained = True
            print(f"AnomalyDetector trained on {len(historical_data)} samples.")
        else:
            print("No historical data provided for training AnomalyDetector.")

    def detect_anomaly(self, data_point: dict):
        """
        Detects anomalies in a single incoming data point.
        Returns an anomaly score and a boolean flag.
        """
        if not self.trained:
            # print("AnomalyDetector not trained yet. Skipping detection.")
            return None, False

        df_point = pd.DataFrame([data_point])
        X_point = df_point[self.features_for_detection]

        # IsolationForest returns -1 for anomalies, 1 for normal
        prediction = self.model.predict(X_point)[0]
        score = self.model.decision_function(X_point)[0] # Lower score indicates higher anomaly likelihood

        is_anomaly = (prediction == -1)
        return score, is_anomaly


# --- 3. Conceptual Forecaster class ---

class Forecaster:
    def __init__(self, order=(1, 1, 1), seasonal_order=(1, 1, 0, 24), exog_features=None):
        """
        Initializes the Forecaster with SARIMAX model parameters.
        'seasonal_order' (s=24) implies 24 15-minute intervals in 6 hours for a daily pattern.
        If daily patterns are over 24 hours, s would be 96 (24 hours * 4 15-min intervals).
        Let's assume a strong daily seasonality at 24 intervals (6 hours) for simplicity.
        """
        self.order = order
        self.seasonal_order = seasonal_order
        self.exog_features = exog_features if exog_features is not None else []
        self.model_fit = None

    def train(self, historical_data: pd.DataFrame):
        """
        Trains the SARIMAX model on historical energy usage data.
        Exogenous variables (weather conditions) are included if specified.
        This would typically be done on a scheduled basis (e.g., daily).
        """
        if historical_data.empty:
            print("No historical data provided for Forecaster training.")
            return

        # Ensure the index is datetime and has a frequency for SARIMAX
        historical_data = historical_data.set_index('timestamp').asfreq('15min')
        y = historical_data['energy_usage']
        exog_data = None
        if self.exog_features and not historical_data[self.exog_features].empty:
            exog_data = historical_data[self.exog_features]

        try:
            model = SARIMAX(y, order=self.order, seasonal_order=self.seasonal_order, exog=exog_data,
                            enforce_stationarity=False, enforce_invertibility=False)
            self.model_fit = model.fit(disp=False)
            print(f"Forecaster trained on {len(historical_data)} samples.")
        except Exception as e:
            print(f"Error training Forecaster: {e}")
            self.model_fit = None

    def forecast(self, steps: int, future_exog_data: pd.DataFrame = None):
        """
        Generates a short-term forecast for the next 'steps' intervals.
        Requires future exogenous variables if the model was trained with them.
        """
        if self.model_fit is None:
            print("Forecaster not trained yet. Cannot generate forecast.")
            return None, None

        if self.exog_features and future_exog_data is None:
            print("Warning: Model trained with exogenous variables, but no future exog data provided for forecast.")
            # Attempt to forecast without exog data (might lead to poor results if exog was crucial)

        try:
            forecast_result = self.model_fit.get_forecast(steps=steps, exog=future_exog_data)
            forecast_mean = forecast_result.predicted_mean
            forecast_conf_int = forecast_result.conf_int()
            return forecast_mean, forecast_conf_int
        except Exception as e:
            print(f"Error generating forecast: {e}")
            return None, None


# --- 4. Main Simulation Loop ---

def main_workflow_simulation():
    print("Starting real-time energy monitoring workflow simulation...")

    # Initialize historical data store (e.g., a DataFrame) for a single region
    # In a real system, this would be a database connection or data lake
    history_duration_days = 7 # Keep 7 days of historical data for training
    start_time = datetime.now() - timedelta(days=history_duration_days)
    current_simulation_time = start_time

    historical_data_list = []
    # Generate some initial historical data
    for _ in range(history_duration_days * 24 * 4): # 15-min intervals per day
        historical_data_list.append(simulate_streaming_data(start_time, current_simulation_time, region_id=1))
        current_simulation_time += timedelta(minutes=15)
    historical_df = pd.DataFrame(historical_data_list)
    historical_df = historical_df.set_index('timestamp').asfreq('15min').reset_index()

    print(f"Initialized with {len(historical_df)} historical data points.")

    # Initialize Anomaly Detector and Forecaster
    anomaly_detector = AnomalyDetector(model_type='IsolationForest', contamination=0.01)
    # We might use 'temperature' and 'humidity' as exogenous variables for SARIMAX
    forecaster = Forecaster(order=(2, 1, 1), seasonal_order=(1, 1, 0, 96), # s=96 for 24-hour cycle with 15min data
                            exog_features=['temperature', 'humidity'])

    # Train models on initial historical data
    anomaly_detector.train(historical_df)
    forecaster.train(historical_df)

    # Simulation parameters
    total_simulation_hours = 24 * 3 # Simulate for 3 days
    interval = timedelta(minutes=15)
    total_intervals = int(total_simulation_hours * 60 / 15)

    anomaly_check_interval_count = 0
    forecast_generation_interval_count = 0
    retrain_anomaly_after_intervals = 24 * 4 * 7 # Retrain anomaly detector every 7 days (conceptual)
    retrain_forecast_after_intervals = 24 * 4 * 1 # Retrain forecaster daily (conceptual)

    print(f"Simulating for {total_simulation_hours} hours ({total_intervals} intervals)...")
    print("------------------------------------------------------------------------------------------------")

    for i in range(total_intervals):
        current_simulation_time += interval
        new_data_point = simulate_streaming_data(start_time, current_simulation_time, region_id=1)

        # --- Anomaly Detection Step ---
        score, is_anomaly = anomaly_detector.detect_anomaly(new_data_point)
        if is_anomaly:
            print(f"[ANOMALY ALERT] at {new_data_point['timestamp']}: "
                  f"Energy Usage={new_data_point['energy_usage']:.2f}, Score={score:.4f}")
        # else:
        #     print(f"Normal at {new_data_point['timestamp']}: Energy Usage={new_data_point['energy_usage']:.2f}")

        # Add new data point to historical data for future retraining/forecasting
        # In a real system, this would be appended to the database/data lake
        historical_df = pd.concat([historical_df, pd.DataFrame([new_data_point]).set_index('timestamp')], ignore_index=False)
        # Keep only the last 'history_duration_days' for efficiency and relevance
        historical_df = historical_df[historical_df.index >= (current_simulation_time - timedelta(days=history_duration_days))]
        historical_df = historical_df.asfreq('15min') # Ensure frequency after concat

        # --- Forecasting Step (Periodic) ---
        forecast_generation_interval_count += 1
        if forecast_generation_interval_count % retrain_forecast_after_intervals == 0:
            print(f"\n[FORECASTING] Generating new forecast at {new_data_point['timestamp']}...")
            # Retrain forecaster with updated historical data
            forecaster.train(historical_df)

            # Generate future exogenous data for the forecast
            forecast_steps = 96 # Forecast next 24 hours (96 15-min intervals)
            future_forecast_times = pd.date_range(start=current_simulation_time + interval, periods=forecast_steps, freq='15min')
            # Simulate future weather conditions (e.g., from a weather forecast API)
            future_exog_data_list = []
            for ft in future_forecast_times:
                future_exog_data_list.append({
                    'timestamp': ft,
                    'temperature': 20 + 5 * np.sin(ft.timetuple().tm_yday * 2 * np.pi / 365) + np.random.normal(0, 0.5),
                    'humidity': 60 + 10 * np.sin(ft.timetuple().tm_yday * 2 * np.pi / 365) + np.random.normal(0, 1)
                })
            future_exog_df = pd.DataFrame(future_exog_data_list).set_index('timestamp')
            future_exog_df = future_exog_df[forecaster.exog_features]

            forecast_mean, forecast_ci = forecaster.forecast(forecast_steps, future_exog_df)
            if forecast_mean is not None:
                print(f"  Forecast for next {forecast_steps} intervals:")
                print(forecast_mean.head(5))
                print(f"  (and up to {forecast_mean.index[-1].strftime('%Y-%m-%d %H:%M')})")
            print("------------------------------------------------------------------------------------------------")

        # --- Model Retraining (Periodic) for Anomaly Detector ---
        anomaly_check_interval_count += 1
        if anomaly_check_interval_count % retrain_anomaly_after_intervals == 0:
            print(f"\n[RETRAINING] Retraining Anomaly Detector at {new_data_point['timestamp']}...")
            anomaly_detector.train(historical_df)
            print("------------------------------------------------------------------------------------------------")

    print("Simulation finished.")


# --- How this solution helps business decisions or operations ---

# 1. Real-time Anomaly Detection:
#    - Operational Efficiency: Immediate alerts on abnormal energy usage can flag equipment malfunctions,
#      leakages, or even theft, allowing for quick investigation and mitigation, reducing downtime and costs.
#    - Safety: Sudden spikes could indicate dangerous overloads, prompting safety checks.
#    - Fraud Detection: Unusually low or high consumption patterns might indicate tampering.

# 2. Short-term Forecasting:
#    - Resource Allocation: Optimize power generation and distribution, ensuring sufficient supply without overproduction.
#      This reduces fuel costs and environmental impact.
#    - Market Trading: For companies involved in energy trading, accurate short-term forecasts can inform buying/selling decisions.
#    - Maintenance Scheduling: Predict peak demand times to schedule non-critical maintenance during off-peak hours.
#    - Grid Stability: Helps grid operators manage load balancing and prevent blackouts.

# 3. Validation and Monitoring:
#    - Ensures model reliability over time, adapting to changing patterns (e.g., population growth, new industries).
#    - Identifies model drift or performance degradation, triggering retraining or model updates, maintaining high accuracy.

# Overall, this integrated workflow provides proactive insights, enabling the power grid company to enhance reliability,
# optimize operations, reduce costs, and improve customer satisfaction.


# Run the simulation
# main_workflow_simulation()

print("Conceptual workflow code defined. To run the simulation, uncomment `main_workflow_simulation()` call.")
```
